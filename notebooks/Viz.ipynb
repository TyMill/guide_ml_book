```python
import pandas as pd

# Example data
data = pd.DataFrame({
    'x': np.random.rand(50),
    'y': np.random.rand(50),
    'category': np.random.choice(['A', 'B', 'C'], 50)
})

# Creating a scatter plot
sns.scatterplot(x='x', y='y', hue='category', data=data)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Seaborn')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color='red', linestyle='--', marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Line Plot')
plt.show()
```

```python
# Example data
y2 = np.cos(x)

plt.plot(x, y, label='Sine Wave')
plt.plot(x, np.cos(x), label='Cosine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Legends')
plt.legend()
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Annotation')
plt.annotate('Max Point', xy=(np.pi/2, 1), xytext=(np.pi/2, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi=100)
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Custom Size and Resolution')
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].apply(lambda x: iris.target_names[x])
print(df.head())
```

```python
plt.hist(df['sepal length (cm)'], bins=20, color='blue', edgecolor='black')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
import seaborn as sns

# Data remains the same (df)

sns.pairplot(df, hue='species')
plt.savefig('pairplot.png')
plt.show()
```

```python
sns.boxplot(x='species', y='petal length (cm)', data=df)
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.title('Box Plot of Petal Length by Species')
plt.show()
```
